# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from decimal import Decimal
sns.set_theme()
```

# Variables

```python
#variables values taken from the pdf
n_h = 0.1
x = 0.1
nu_0 = 2.46607e+15
eta = 6.265e+8
f = 0.4164
g_0 = 2
z = 2
m_e = 9.11e-28
c = 3e+10
e = 4.8e-10
d_list = [10e+14, 10e+18, 10e+21]
pi = 3.14159
```

# Point 1.

The intensity $I$ of light of a given wavelength $\lambda$ is given by this equation:

$I(\lambda) = e^{-\alpha(\nu)d}$

where the frequency $\nu$ is related to $\lambda$ by $\nu = \frac{c}{\lambda}$. The thickness of the slab is $d$. The absorption coefficient \alpha(\nu) is given by the following equation:

$\alpha(\nu) = \frac{e^2f n_H}{4 \pi m_ec}\frac{(1-x)g_0}{\mathcal{Z}}\frac{\Gamma}{(\nu - \nu_0)^2+(\Gamma/4\pi)^2}$

In this formula is not clear if $\Gamma$ is divided by $4 \pi$ or by $4$ and then multiplied for $\pi$. In my implementation, I consider the first option.

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize = (30,7))
y_limit = [(-0.05, 1.05), (-0.05,1.05), (-0.05,1.05)]
for indx, d in enumerate(d_list):
  lamb = np.linspace(1e-8, 1400 * 1e-8, num = 1000000) #linspace between 800 and 1400
  nu = c/lamb
  alpha = (e**2 * f * n_h)/(4*pi*m_e*c) * ((1-x)*g_0)/z * eta/((nu-nu_0)**2 + (eta/(4*pi))**2)
  I = np.exp(-alpha * d)
  axs[indx].set_title(f"Spectrum d = {Decimal(d):.2E}")
  axs[indx].plot(lamb, I, c=np.random.rand(3,))
  axs[indx].set_xlabel("Wavelength")
  axs[indx].set_ylabel("Intensity")
  axs[indx].set_ylim(y_limit[indx])
  indx+=1
```

The difference in the thickness parameter $d$ affects the amount of radiation absorbed. This variation results in a greater density of atoms (and consequently of electrons) which leads to greater absorption.
This phenomenon is particularly evident in the last two graphs, where the absorption peaks are greater.
